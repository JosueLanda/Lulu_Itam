```python
import subprocess
from subprocess import Popen, PIPE

subprocess.run(["gcc", "./sistema_ecuaciones.c", "-lm","-o", "executable"])
```

```python
for i in range(2,10):
    for j in range(3,10):
        for k in range(4,5):

            exe = ("./executable")
            with Popen(exe, shell=True, stdin=PIPE,stdout=PIPE) as proc:
                
                output, _ = proc.communicate(bytes(f'{j} {i} {k} {k} {j} {i}',encoding='utf-8'))
                
                if str(output).find('inf')==-1:
                    print(f'{j} {i} {k}')
                    print(f'{k} {j} {i}')
                    print(str(output).replace("b","").replace("'",""))
                    print("\n\n")
            
        if j == 7:
            break
```
